# Fourier Series

$
f(t)=\frac{a_0}{2}+\sum_{n=1}^{\infty}\left[a_n\cos(n\omega_0t)+b_n\sin(n\omega_0t)\right]
$

**Where:**

* $f(t$) – Periodic signal.
* $a_0$ – DC (average) component.
* $a_n$ – Cosine coefficients (even part).
* $b_n$ – Sine coefficients (odd part).
* (n) – Harmonic number ((1,2,3,\ldots)).
* $\omega_0=\dfrac{2\pi}{T}$ – Fundamental angular frequency (rad/s).
* $T$ – Fundamental period.




**Fourier coefficients:**

$
a_0=\frac{2}{T}\int_{t_0}^{t_0+T}f(t),dt
$

*Average (DC) value of the signal.*

$
a_n=\frac{2}{T}\int_{t_0}^{t_0+T}f(t)\cos(n\omega_0t),dt
$

*Coefficient of the $n^{\text{th}}$cosine harmonic.*

$
b_n=\frac{2}{T}\int_{t_0}^{t_0+T}f(t)\sin(n\omega_0t),dt
$

*Coefficient of the $n^{\text{th}}$ sine harmonic.*
