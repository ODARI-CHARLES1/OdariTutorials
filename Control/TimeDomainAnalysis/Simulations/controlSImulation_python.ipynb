```python
import numpy as np
from control import matlab as mt
from matplotlib import pyplot as plt
import control as ct
import scienceplots
import seaborn as sns
import warnings 

warnings.filterwarnings("ignore")
plt.style.use(["science","no-latex"])
```

```python
num=[25]
den=[1,6,25]
sys=mt.tf(num,den)
sys
```

```python
yout,t=mt.step(sys)
plt.figure(figsize=(8,4),facecolor="white")
plt.plot(t,yout,label="System Response $(c(t)$",color="#0072b2",linewidth=2.5)
plt.axhline(y=1.0,color="r",linestyle="--",linewidth=1.5,label="Steady-Statue Value")
plt.title('Step Response Verification (Python)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Time (seconds)', fontsize=12)
plt.ylabel('Amplitude', fontsize=12)
plt.xlim(0, 2.5)
plt.ylim(0, 1.3)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(loc='lower right', fontsize=11)
plt.show()
```
